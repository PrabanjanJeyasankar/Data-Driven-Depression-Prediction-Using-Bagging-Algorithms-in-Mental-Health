```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
var = ['gender','who_bmi','depression_severity','depressiveness','suicidal','depression_diagnosis','depression_treatment','anxiety_severity','anxiousness','anxiety_treatment','anxiety_diagnosis','sleepiness']
  
for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```
